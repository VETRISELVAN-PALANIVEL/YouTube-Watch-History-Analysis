```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_json(r"C:\Users\Admin\Downloads\VETRI\DATA ANALYTICS\Extracted files\Utube history\YouTube and YouTube Music\history\watch-history.json")
data['time']=pd.to_datetime(data['time'],format='mixed',errors='coerce')
data.head()
```

```python
data['hour']=data['time'].dt.hour
data['month']=data['time'].dt.month_name()
data['day']=data['time'].dt.day_name()
data.head()
```

```python
data.shape
```

```python
data['month'].value_counts().plot(kind='bar',color='red')
```

```python
data['hour'].value_counts().sort_index().plot(kind='bar',color='violet')
plt.ylabel("No.of videos watched")
plt.title("YouTube Viewing Frequency by Hour")
```

```python
data['day'].value_counts().plot(kind='bar',color='black')
plt.xlabel('day',color='blue')
plt.ylabel('No.of videos watched',color='blue')
plt.title('Viewing Frequency by Day of the Week',color='green')
```

```python
data['title'].value_counts().head(10).plot(kind='bar',color='orange')
plt.xlabel('Titles',color='brown')
plt.ylabel('No.of times watched',color='brown')
```

```python
# Extract day of the week and hour from the 'time' column
data['time'] = pd.to_datetime(data['time'])  # Ensure 'time' is in datetime format
data['day'] = data['time'].dt.day_name()
data['hour'] = data['time'].dt.hour

# Frequency by day of the week
day_counts = data['day'].value_counts()
print("Watch Frequency by Day:")
print(day_counts)

day_counts.plot(kind='bar', color='orange')
plt.title('YouTube Viewing Frequency by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Watch Count')
plt.show()

# Frequency by hour of the day
hour_counts = data['hour'].value_counts().sort_index()
print("Watch Frequency by Hour:")
print(hour_counts)

hour_counts.plot(kind='bar', color='green')
plt.title('YouTube Viewing Frequency by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Watch Count')
plt.show()
```
